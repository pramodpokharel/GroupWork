```python
import pandas as pd
path = (r'......xlsx')

data_all = pd.read_excel(path)

round(data_all.describe(),1)
```

Compute the central tendency measures

```python
import numpy as np
import statistics


for col_name, data in data_all.items():
    print(f'\n** Sensor {col_name} **')
    print(f'min, max     = {np.min(data,axis=0):.1f}, {np.max(data,axis=0):.1f}')
    print(f'mean, median = {np.mean(data,axis=0):.1f}, {np.median(data,axis=0):.1f}')
    print(f'std, var     = {np.std(data,axis= 0):.1f}, {np.var(data,axis= 0):.1f}')
    print(f' mode   =   {statistics.mode(data)}')

```
