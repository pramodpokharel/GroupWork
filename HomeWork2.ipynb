```python
##Generate a random sample of 50 observations for each sensor

S1_sample = np.random.choice(df['S1'], size=50)
s2_sample = np.random.choice(df['S2'], size=50)
S3_sample = np.random.choice(df['S3'], size=50)
```

```python
# Compute upper and lower bounds with 90%, 95%, 99% confidence intervals for each sensor
s1_mean = S1_sample.mean()
s1_std = S1_sample.std()

s2_mean = S2_sample.mean()
s2_std = S2_sample.std()

s3_mean = S3_sample.mean()
s3_std = S3_sample.std()



# For 90% confidence interval

##sample size = 50
z_score_90 = 1.645

s1_upper_90 = s1_mean + z_score_90 * s1_std / np.sqrt(50)
s1_lower_90 = s1_mean - z_score_90 * s1_std / np.sqrt(50)
print('Lower bound of S1 sample for confidence intervals 90% = ',s1_lower_90)
print('Upper bound of S1 sample for confidence intervals 90% = ',s1_upper_90,'\n')


s2_upper_90 = s2_mean + z_score_90 * s2_std / np.sqrt(50)
s2_lower_90 = s2_mean - z_score_90 * s2_std / np.sqrt(50)
print('Lower bound of S2 sample for confidence intervals 90% = ',s2_lower_90)
print('Upper bound of S2 sample for confidence intervals 90% = ',s2_upper_90,'\n')

s3_upper_90 = s3_mean + z_score_90 * s3_std / np.sqrt(50)
s3_lower_90 = s3_mean - z_score_90 * s3_std / np.sqrt(50)
print('Lower bound of S3 sample for confidence intervals 90% = ',s3_lower_90)
print('Upper bound of S3 sample for confidence intervals 90% = ',s3_upper_90,'\n')


# For 95% confidence interval
z_score_95 = 1.96

s1_upper_95 = s1_mean + z_score_95 * s1_std / np.sqrt(50)
s1_lower_95 = s1_mean - z_score_95 * s1_std / np.sqrt(50)
print('Lower bound of S1 sample for confidence intervals 95% = ',s1_lower_95)
print('Upper bound of S1 sample for confidence intervals 95% = ',s1_upper_95,'\n')


s2_upper_95 = s2_mean + z_score_95 * s2_std / np.sqrt(50)
s2_lower_95 = s2_mean - z_score_95 * s2_std / np.sqrt(50)
print('Lower bound of S2 sample for confidence intervals 95% = ',s2_lower_95)
print('Upper bound of S2 sample for confidence intervals 95% = ',s2_upper_95,'\n')


s3_upper_95 = s3_mean + z_score_95 * s3_std / np.sqrt(50)
s3_lower_95 = s3_mean - z_score_95 * s3_std / np.sqrt(50)
print('Lower bound of S3 sample for confidence intervals 95% = ',s3_lower_95)
print('Upper bound of S3 sample for confidence intervals 95% = ',s3_upper_95,'\n')

# For 99% confidence interval
z_score_99 = 2.576

s1_upper_99 = s1_mean + z_score_99 * s1_std / np.sqrt(50)
s1_lower_99 = s1_mean - z_score_99 * s1_std / np.sqrt(50)
print('Lower bound of S1 sample for confidence intervals 99% = ',s1_lower_99)
print('Upper bound of S1 sample for confidence intervals 99% = ',s1_upper_99,'\n')


s2_upper_99 = s2_mean + z_score_99 * s2_std / np.sqrt(50)
s2_lower_99 = s2_mean - z_score_99 * s2_std / np.sqrt(50)
print('Lower bound of S2 sample for confidence intervals 99% = ',s1_lower_99)
print('Upper bound of S2 sample for confidence intervals 99% = ',s1_upper_99,'\n')


s3_upper_99 = s3_mean + z_score_99 * s3_std / np.sqrt(50)
s3_lower_99 = s3_mean - z_score_99 * s3_std / np.sqrt(50)
print('Lower bound of S3 sample for confidence intervals 99% = ',s1_lower_99)
print('Upper bound of S3 sample for confidence intervals 99% = ',s1_upper_99)

```
